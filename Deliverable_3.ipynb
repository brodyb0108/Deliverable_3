```python
#(1)
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#(2)
iris = sns.load_dataset('iris')
iris.head()
```

```python
#(3)
iris = iris.drop(labels = 'species' , axis = 1)
iris.head()
```

```python
#(4)
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
iris.head()
```

```python
#(4)
iris['petal_sum'] = iris['petal_length'] + iris['petal_width']
iris.head()
```

```python
#(5)
mean = [np.mean(iris['sepal_length'])], [np.mean(iris['sepal_width'])], [np.mean(iris['sepal_sum'])], [np.mean(iris['petal_length'])], [np.mean(iris['petal_width'])], [np.mean(iris['petal_sum'])]
summary_dict = {'mean' :mean}
summary_df = pd.DataFrame(data = summary_dict)
summary_df
```

```python
#(5)
std = [np.std(iris['sepal_length'])], [np.std(iris['sepal_width'])], [np.std(iris['sepal_sum'])], [np.std(iris['petal_length'])], [np.std(iris['petal_width'])], [np.std(iris['petal_sum'])]
summary_dict = {'std' :std}
summary_df = pd.DataFrame(data = summary_dict)
summary_df
```

```python
#(5)
min = [np.min(iris['sepal_length'])], [np.min(iris['sepal_width'])], [np.min(iris['sepal_sum'])], [np.min(iris['petal_length'])], [np.min(iris['petal_width'])], [np.min(iris['petal_sum'])]
summary_dict = {'min' :min}
summary_df = pd.DataFrame(data = summary_dict)
summary_df
```

```python
#(5)
max = [np.max(iris['sepal_length'])], [np.max(iris['sepal_width'])], [np.max(iris['sepal_sum'])], [np.max(iris['petal_length'])], [np.max(iris['petal_width'])], [np.max(iris['petal_sum'])]
summary_dict = {'max' :max}
summary_df = pd.DataFrame(data = summary_dict)
summary_df
```
